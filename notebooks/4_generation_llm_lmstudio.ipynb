# 4 - Use small local LLMs with LM Studio

LMStudio is a platform that allows users to build, deploy, and manage machine learning models and applications locally on their own machines. It provides tools and interfaces for data scientists and developers to create AI-powered solutions without relying on cloud services.

More info at: https://lmstudio.ai/docs/app/basics

```python
import requests

# URL of the local LLM server exposing an OpenAI-compatible API such as LMStudio
# In this example, it is assumed that a model (e.g., gemma-3-4b) is running on port 1234.
url = 'http://localhost:1234/v1/chat/completions'

# Payload following the OpenAI Chat Completions API format.
# "model": name of the model served by the local backend.
# "messages": list of messages to simulate a chat conversation.
payload = {'model':'gemma-3-4b',
           'messages': [
               {'role':'user','content':'Explique le S&OP en français de manière simple et concise.'}
               ]
          }

# Send an HTTP POST request to the LLM endpoint with the payload as JSON.
resp = requests.post(url, json=payload)

# Print the JSON response returned by the model (contains the generated text, metadata, etc.).
print(resp.json())
```
