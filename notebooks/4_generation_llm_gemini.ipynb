# 4 - Use Gemini API

Get a free API key from: https://aistudio.google.com/app/api-keys

Store your API key in a `.env` file as `GOOGLE_API_KEY=your_api_key`

14k queries limit per day: 
* `gemma-3-4b`
* `gemma-3-12b`
* `gemma-3-27b` 

20 queries limit per day: 
* `gemini-2.5-flash`
* `gemini-3-flash`

```python
import os
from dotenv import load_dotenv

# Load variables from the .env file
load_dotenv()

# Retrieve Google API key from environment variables
API_KEY = os.getenv("GOOGLE_API_KEY")
```

```python
from google import genai

client = genai.Client(api_key=API_KEY)

response = client.models.generate_content(
    model="gemma-3-12b-it",
    contents="Explique le S&OP en français de manière simple et concise.",
)

print(response.text)
```
